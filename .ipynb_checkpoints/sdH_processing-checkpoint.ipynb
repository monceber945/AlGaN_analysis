We start with importing data from csv. Data was saved as **(field,T ; voltave, V)**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

data4K = pd.read_csv('data/sdH4K.csv',header=None)
data4K.columns = ['field', 'voltage']

data10K = pd.read_csv('data/sdH10K.csv',header=None)
data10K.columns = ['field', 'voltage']


```

```python
#Current = 50nA for 4K. 100nA for 10K. creating a resistance column
data4K['resistance']= data4K['voltage']/(50* 10**(-9))
data10K['resistance']= data10K['voltage']/(100* 10**(-9))


data4K['dR/R'] = (data4K['resistance']-data4K['resistance'][0])/data4K['resistance'][0]
data10K['dR/R'] = (data10K['resistance']-data10K['resistance'][0])/data10K['resistance'][0]
```

```python
sdH4Kgraph = px.scatter(data4K, x = data4K['field'], y = data4K['dR/R'], width = 1000, template = 'simple_white')
sdH4Kgraph.update_layout(
    title=r'4K Mesa sdH oscillations',
    xaxis_title=r'Magnetic field, T',
    yaxis_title=r'$\frac{R_i-R(0)}{R(0)}$', font=dict(
        size=13
    ))
sdH4Kgraph.update_traces(marker=dict(size=4,opacity=0.8, color = "#7AC5CD",
                              line=dict(width=1,
                                        color="#53868B")),
                  selector=dict(mode='markers'))

sdH4Kgraph.show()


```
